```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_excel("TESTE_0912-1412.xlsx", parse_dates=["Data e hora"]) 

fig, ax =plt.subplots()

ax.plot(dados['Data e hora'],dados['T1'])
#ax.plot(dados['Data e hora'],dados['U1'])
n = 10
step = len(dados) // n
xticks = dados['Data e hora'][::step]

plt.gca().set_xticks(xticks)
plt.gca().set_xticklabels(xticks, rotation=90)
```
